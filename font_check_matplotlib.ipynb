````python
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams.update({
    "text.usetex": True,   # ask Matplotlib to use system LaTeX
    "font.family": "serif",
    "font.serif": ["Times New Roman", "Times"],
    "ps.fonttype": 42,
    "pdf.fonttype": 42,
})

plt.plot([0,1],[0,1])
plt.xlabel(r'Test $\alpha+\beta=\gamma$')
plt.ylabel(r'Times text with math')
plt.title(r'LaTeX check: $\int_0^1 x^2 dx$')
plt.savefig("latex_test.pdf")
plt.show()
````

````python
from matplotlib import font_manager as fm
from matplotlib.font_manager import FontProperties

# List all fonts Matplotlib sees (full paths)
all_fonts = fm.findSystemFonts(fontpaths=None, fontext='ttf')
len(all_fonts), all_fonts[:10]  # peek

# Search for specific families
targets = ["Times New Roman", "Arial", "Helvetica", "Nimbus Sans", "Liberation Sans"]
for name in targets:
    try:
        path = fm.findfont(FontProperties(family=name), fallback_to_default=False)
        print(f"{name} -> {path}")
    except Exception as e:
        print(f"{name} -> NOT FOUND ({e})")
````
